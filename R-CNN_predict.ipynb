```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

IMG_ROW_SIZE = 384
IMG_COL_SIZE = 512
```

```python

model = tf.keras.models.load_model("model1.keras")
```

```python

image = cv2.imread("test_images/timage4.jpeg")

image = cv2.resize(image, (IMG_COL_SIZE, IMG_ROW_SIZE))
image = image / 255.0
image = np.expand_dims(image, axis=0)
```

```python

pred = model.predict(image)
mask = (pred[0] > 0.5).astype("uint8")

plt.figure(figsize=(5,10))
plt.imshow(image[0], cmap="gray")
plt.axis("off")
plt.show()

plt.figure(figsize=(5,10))
plt.imshow(pred[0], cmap="gray")
plt.axis("off")
plt.show()

plt.figure(figsize=(5,10))
plt.imshow(mask, cmap="gray")
plt.axis("off")
plt.show()
```
